```python
spark.read.table("samples.nyctaxi.trips")
```

```python
spark.sql(f"""create or replace table {catalog}.{schema}.transactions_summary as select product,paymentMethod,
           sum(quantity) as total_quantity,
           sum(totalPrice) as total_price
           from samples.bakehouse.sales_transactions group by product,paymentMethod """)
```
